# Q1: Algorithm Analysis

## Background on Algorithm Analysis (50 points total)

- Constant Time (O(1)): Regardless of the input size, the algorithm takes a constant amount of time to complete.
    - Accessing an element in an array by index.
    - Inserting or deleting an element at the beginning of a linked list.


- Linear Time (O(n)): The runtime grows linearly with the size of the input.
    - Iterating through a list or array once.
    - Searching for an element in an unsorted list by traversing it.


- Logarithmic Time (O(log n)): As the input size grows, the runtime increases logarithmically.
    - Binary search in a sorted array.
    - Operations in a balanced binary search tree (BST).
    

- Quadratic Time (O(n^2)): The runtime grows quadratically with the input size.
    - Nested loops where each loop iterates through the input.
    - Some sorting algorithms like Bubble Sort or Selection Sort.


- Exponential Time (O(2^n)): The runtime doubles with each addition to the input size.
    - Algorithms involving recursive solutions that make repeated calls.
    - Solving the Traveling Salesman Problem using brute force.
    

- Factorial Time (O(n!)): The runtime grows factorially with the input size.
    - Permutation problems that explore all possible combinations, like the brute force solution for the Traveling Salesman Problem with no optimizations.

## Bubble Sort (Example)

Use the below analysis and explanation as a template for answering the following questions on algorithm analysis. You are to analyze each algorithm line by line, calculate the runtime complexity, and provide an explanation in markdown.

```python
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):  # O(n)
        for j in range(0, n-i-1):  # O(n)
            if arr[j] > arr[j+1]:  # O(1)
                arr[j], arr[j+1] = arr[j+1], arr[j]  # O(1)

# Total Runtime Complexity:
# Big O notation: O(n^2)
# Big Omega notation: Ω(n)
# Big Theta notation: Θ(n^2)
```

Explanation:

- The outer loop iterates 'n' times, where 'n' is the length of the array.
- The inner loop also iterates 'n' times in the worst case.
- The comparisons and swapping operations inside the inner loop are constant time ('O(1)').

This results in a total runtime complexity of O(n^2) in the worst case, Ω(n) in the best case (already sorted), and Θ(n^2) in the average and worst cases, as both loops iterate over the input array.

## 1A Merge Sort (10 points)

```python
def merge_sort(arr):
    if len(arr) > 1:  # O(1)
        mid = len(arr) // 2  # O(1)
        left_half = arr[:mid] # O(n)
        right_half = arr[mid:]  # O(n)

        merge_sort(left_half)  # O(log n)
        merge_sort(right_half)  # O(log n)

        i = j = k = 0  # O(1)

        while i < len(left_half) and j < len(right_half):  # O(n)
            if left_half[i] < right_half[j]:  # O(1)
                arr[k] = left_half[i]  # O(1)
                i += 1 # O(1)
            else:
                arr[k] = right_half[j]  # O(1)
                j += 1 # O(1)
            k += 1 # O(1)

        while i < len(left_half):  # O(n)
            arr[k] = left_half[i]  # O(1)
            i += 1 # O(1)
            k += 1 # O(1)

        while j < len(right_half):   # O(n)
            arr[k] = right_half[j]   # O(1)
            j += 1 # O(1)
            k += 1 # O(1)

# Total Runtime Complexity:
# Big O notation: O(n log n)
# Big Omega notation: Ω(n log n)
# Big Theta notation Θ(n log n)
```

Explanation:
- Repeatedly divides the array into two halves, leading to log(n) levels of recursion, where 'n' is the size of the input.
- Processes, compares, and combines all elements during each level of recursion (O(n)).
- Individually compares and assigns values (O(1)) for each element 'n' (O(n)).

Worst, best, and average case complexity is O(n log n), since the merging step remains consistent at every level of recursion.

## 1B Binary Search (10 points)

```python
def binary_search(arr, target):
    low = 0  # O(1)
    high = len(arr) - 1  # O(1)

    while low <= high:  # O(log n)
        mid = (low + high) // 2  # O(1)
        mid_val = arr[mid]  # O(1)

        if mid_val == target:  # O(1)
            return mid  # O(1)
        elif mid_val < target:  # O(1)
            low = mid + 1 # O(1)
        else: # O(1)
            high = mid - 1  # O(1)

    return -1  # O(1)

# Total Runtime Complexity:
# Big O notation: O(log n)
# Big Omega notation: Ω(1)
# Big Theta notation: Θ(log n)
```

Explanation:
- Repeatedly cuts the search range in half, (O(log n)), where 'n' is the total number of elements.
- Picks the middle element, checks if it is the target, and adjusts the search range if value is not found (O(1)).
- Shifts the search range by adjusting the low and high pointers.

Worst and average case takes O(log n) time, since the pattern is consistent across searches. Best case is Ω(1) if the target is found immediately.

## 1C Linear Search (10 points)

```python
def linear_search(arr, target):
    for i in range(len(arr)): # O(n)
        if arr[i] == target:  # O(1)
            return i   # O(1)
    return -1  # O(1) 

# Total Runtime Complexity:
# Big O notation: O(n)
# Big Omega notation: Ω(1)
# Big Theta notation: Θ(n)
```

Explanation:
- Scans through the array one element at a time, making 'n' checks, where n is the number of elements.
- If the target is found early, the loop immediately exits.

Worst and average case is O(n) since most times involve searching halfway through the list. The best case is Ω(1), when the first element matches.

## 1D Selection Sort (10 points)

```python
def selection_sort(arr):
    n = len(arr) # O(1)
    for i in range(n - 1):  # O(n)
        min_index = i   # O(1)
        for j in range(i + 1, n):   # O(n)
            if arr[j] < arr[min_index]:   # O(1)
                min_index = j   # O(1)
        arr[i], arr[min_index] = arr[min_index], arr[i]   # O(1)

# Total Runtime Complexity:
# Big O notation:  O(n^2)
# Big Omega notation:  Ω(n^2)
# Big Theta notation:  Θ(n^2)
```

Explanation:
- The outer loop runs (n - 1) times, where 'n' is the number of items in the set.
- The inner loop scans the remaining unsorted elements to find the minimum (O(n)).
- Swaps elements once per iteration of the outer loop.

Worst-case, best-case, and average-case complexity is O(n^2), since it always scans the full remaining list, even if the array is already sorted.

# 1E Heap Sort (10 points)

```python
def heapify(arr, n, i):
    largest = i   # O(1)                     
    left = 2 * i + 1    # O(1)               
    right = 2 * i + 2     # O(1)         

    if left < n and arr[left] > arr[largest]: # O(1)
        largest = left # O(1)

    if right < n and arr[right] > arr[largest]: # O(1)
        largest = right # O(1)

    if largest != i: # O(1)
        arr[i], arr[largest] = arr[largest], arr[i] # O(1)
        heapify(arr, n, largest) # O(log n)


def heap_sort(arr):
    n = len(arr) # O(1)

    for i in range(n // 2 - 1, -1, -1): # O(n)
        heapify(arr, n, i) # O(log n)

    for i in range(n - 1, 0, -1): # O(n)
        arr[i], arr[0] = arr[0], arr[i]   # O(1)
        heapify(arr, i, 0)   # O(log n)             
```

Explanation:
- Builds a heap, working from the bottom up (O(n)).
- Largest element swaps places with the last element in the heap. Heapify fixes the heap 'n' times, where 'n' is the number of times it occurs as the array gets sorted.
- Running heapify moves elements down the heap (O(log n)).

Worst-case, best-case, and average-case is O(n log n), since every step follows the same heapify process.

```python
import math

def is_prime(n):
    if n <= 1:  # O(1)
        return False  # O(1)
    for i in range(2, int(math.sqrt(n)) + 1):  # O(sqrt(n))
        if n % i == 0:  # O(1)
            return False  # O(1)
    return True  # O(1)

# Total Runtime Complexity:
# Big O notation: O(sqrt(n))
# Big Omega notation: Ω(1)
# Big Theta notation: Θ(sqrt(n))
```

Explanation:
- Checks if 'n' is <=1, returns False if it is.
- If n is 2, it immediately returns True since 2 is the only even prime.
- If n is even and greater than 2, it returns False.
- The loop then starts at 3 and only checks odd numbers until it reaches its square root.

Worst and average-case is O(sqrt(n)) time, since it checks divisibility up to the square root. The best case is Ω(1), where the function exits early.

## Plotting the Difference

```python
from matplotlib import pyplot as plt
from scipy.stats import norm

# Plot histogram for each numeric column
for column_name in numeric_df.columns:
	data = numeric_df[column_name].dropna()

	# calculate actual mean and variance for this column
	actual_mean = np.mean(data) # calculate mean
	actual_variance = np.var(data, ddof=1) # ddof = 1 to calculate sample variance

	# plot histogram of the data
	plt.hist(data, bins=30, density=True, # density = True to normalize the data
		  alpha=0.8, color='pink', label = 'Actual Data')

	# plot theoretical normal distribution
	xmin, xmax = plt.xlim() # get the limits of x-axis
	x = np.linspace(xmin, xmax, 100) # create 100 points between xmin and xmax
	p = norm.pdf(x, actual_mean, np.sqrt(actual_variance) # pdf = probability density function
			  )
	plt.plot(x, p, 'k', linewidth=2, label='Fitted Normal Distribution') # plot the normal distribution

	# add labels and title
	plt.xlabel(column_name)
	plt.ylabel('Density')
	plt.title(f'Histogram of {column_name} and Fitted Normal Distribution')
	plt.legend(['Theoretical Normal Distribution', 'Actual Data'])
	plt.show()
```
